```python
Journee = []
l=0
for i in range(11):

    for j in range(38):
        for k in range(10):
            if l > (df.shape[0]-1):
                break;
            else:
                l+=1
                Journee.append(j+1)
    
df['Journee']=Journee
```

```python
Season = []
l=0
season = 2009
for i in range(11):
    for j in range(380):
        if l > (df.shape[0]-1):
            break;
        else:
            l+=1
            Season.append(season+i)
    
df['Season']=Season
```

```python
bookmaker_A = "B365A"
bookmaker_D = "B365D"
bookmaker_H = "B365H"
m = 1
gain_brut = []
perte = []
gain_net = []
for i in range(df.shape[0]):
    if df.FTR.iloc[i] == "A":
        gain_brut.append(0)
    elif df.FTR.iloc[i] == "D":
        gain_brut.append(0)
    elif df.FTR.iloc[i] == "H":
        gain_brut.append(df[bookmaker_H].iloc[i] * m * 1)
        
    perte.append(m * 1)
    gain_net.append(gain_brut[-1] - perte[-1])
    
df["Gain_net"] = gain_net
df["Gain_brut"] = gain_brut
df["Perte"] = perte
```

```python
"""Ratio de victoire à domicile"""
i=0
HomeWinRate=[]#Tau de victoire a domicile de l'equipe a domicile (avant le match)
HomeLooseRate=[]#Tau de defaite a domicile de l'equipe a domicile(avant le match)

AwayWinRate=[]#Tau de victoire a l'exterieur de l'equipe a l'exterieur (avant le match)
AwayLooseRate=[]#Tau de defaite a l'exterieur de l'equipe a l'exterieur (avant le match)

HomeScoredGoalsRate=[]#Tau de but marques a domicile par l'equipe a domicile (avant le match)
AwayScoredGoalsRate=[]#Tau de but marques a l'exterieur par l'equipe a l'exterieur (avant le match)

HomeConcededGoalsRate=[]#Tau de but encaisses a domicile par l'equipe a domicile (avant le match)
AwayConcededGoalsRate=[]#Tau de but encaisses a l'exterieur par l'equipe a l'exterieur (avant le match)


while i<df.shape[0]:
    df_ = df.iloc[0:i+1,:]
    HomeWinRate.append(df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,6].value_counts().reindex(['H','A','D']).fillna(0).loc['H']/
                       df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,6].shape[0])
   
    HomeLooseRate.append(df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,6].value_counts().reindex(['H','A','D']).fillna(0).loc['A']/
                         df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,6].shape[0])
    
    AwayWinRate.append(df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,6].value_counts().reindex(['H','A','D']).fillna(0).loc['A']/
                       df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,6].shape[0])
        
    AwayLooseRate.append(df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,6].value_counts().reindex(['H','A','D']).fillna(0).loc['H']/
                       df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,6].shape[0])
    
    
    HomeScoredGoalsRate.append(df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,4].sum()/df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,4].shape[0])
    AwayScoredGoalsRate.append(df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,5].sum()/df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,5].shape[0])
    HomeConcededGoalsRate.append(df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,5].sum()/df_.loc[(df_.HomeTeam == df_.iloc[i,2])].iloc[0:i,5].shape[0])
    AwayConcededGoalsRate.append(df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,4].sum()/df_.loc[(df_.AwayTeam == df_.iloc[i,3])].iloc[0:i,4].shape[0])
        
    i+=1
 
"""Incorporation des variables au DataFrame"""
df['HomeWinRate']=HomeWinRate
df['AwayWinRate']=AwayWinRate
df['HomeLooseRate']=HomeLooseRate
df['AwayLooseRate']=AwayLooseRate
df['HomeScoredGoalsRate']=HomeScoredGoalsRate
df['AwayScoredGoalsRate']=AwayScoredGoalsRate
df['HomeConcededGoalsRate']=HomeConcededGoalsRate
df['AwayConcededGoalsRate']=AwayConcededGoalsRate
```

```python
feature_list = ['HomeTeam','AwayTeam','B365H','B365D','B365A','Log_B365','Journee','Season', "Gain_net",
               'HomeWinRate','AwayWinRate','HomeLooseRate','AwayLooseRate','HomeScoredGoalsRate',
               'AwayScoredGoalsRate','HomeConcededGoalsRate','AwayConcededGoalsRate', "FTR"]
df = df[feature_list]
df.fillna(value=0,inplace=True)
df.to_csv(path_or_buf="C:\\Users\\Utilisateur\\Desktop\\Win_Loose_Goals_Rate_and_Odds_Season_Season1920_.csv",index=False)
```
